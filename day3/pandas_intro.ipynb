# Pandas

Pandas introduces the _dataframe_: 
- Organized data structure
- Stores data in columns
- Great for keeping data clean
    - easy aggregating and plotting

```python
import pandas as pd
```

## Our First DataFrame

```python
xs = np.linspace(0, 10)
y1 = xs + np.random.rand(xs.shape[0])
y2 = np.power(xs, 0.5)
```

Now we have 3 different arrays, with which we can define a _dataframe_

```python
data = {'x':xs, 'y1':y1, 'y2':y2}
dframe = pd.DataFrame(data)
```

```python
dframe.head()
```

##### Accessing Elements

```python
dframe['y1'] ## column selection
```

```python
dframe[['y1', 'y2']]
```

```python
dframe[['y1', 'y2']].to_numpy()
```

```python
dframe.iloc[25:].head() ## row selection
```

**Row and Column**

```python
dframe.iloc[25:][['y1', 'y2']].head()
```

```python
dframe.iloc[25:, 1:3].head()
```

##### Condition Based Subsetting

```python
dframe[dframe['x'] < 5.].head()
```

```python
dframe[(dframe['x'] < 5.) & (dframe['y2'] < 2.)] .head()
```

#### Adding New Columns

```python
dframe['diff'] = dframe['y1'] - dframe['y2']
```

```python
dframe.head()
```

## Sci-Kit Learn Dataset

```python
import sklearn.datasets
iris = sklearn.datasets.load_iris(as_frame=True)
dframe = iris['data']
dframe['type'] = iris['target']
```

![alt text](flower.png "Title")

```python
dframe.head()
```

```python
dframe.columns = ['sepal_l', 'sepal_w', 'petal_l', 'petal_w', 'type']
```

```python
dframe.head()
```

## DataFrames + Seaborn

```python
sns.set_style("white")
sns.color_palette("Greens")
plt.figure(dpi=100)
sns.scatterplot(x='petal_l', y='sepal_l', hue='type',
               data=dframe, palette='bright')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, dpi=100, figsize=(10, 5))
sns.violinplot(x='type', y='sepal_l',
               data=dframe, palette='Greens',
              ax=axes[0])
sns.boxplot(x='type', y='sepal_l',
               data=dframe, palette='Greens',
              ax=axes[1])
plt.show()
```

##### Make an Extra Field

```python
dframe['type2'] = dframe['sepal_w'] > 3.
```

```python
sns.set_style("white")
sns.color_palette("Greens")
plt.figure(dpi=100)
sns.scatterplot(x='petal_l', y='sepal_l', hue='type', style='type2',
               data=dframe, palette='bright')
plt.show()
```
